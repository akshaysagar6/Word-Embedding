```python
import gensim
from gensim.models import Word2Vec,keyedvectors
```

```python
corpus=[x.split() for x in newdf['reviewText'].values]
```

```python
corpus[2]
```

```python
model=gensim.models.Word2Vec(corpus)
```

```python
model.wv.index_to_key
```

```python
model.wv.most_similar('hi')
```

```python
# average word2vec
def avg_word2vec(doc):
    valid_words=[model.wv[word] for word in doc if word in model.wv.key_to_index ]
    if valid_words:
        return np.mean(valid_words,axis=0)
    else:
        # Return a zero vector if no valid words are found
        return np.zeros(model.wv.vector_size)
```

```python
x=[]
for i in range(len(corpus)):
    x.append(avg_word2vec(corpus[i]))
```

```python
x_df=pd.DataFrame(x)
```

```python
trainpredt=nm.predict(xtrain)
print(classification_report(trainpredt,ytrain))
```
