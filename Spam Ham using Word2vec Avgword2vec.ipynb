```python
df=pd.read_csv(r"D:\CognoRise\spam\spam.csv")
df
```

```python
from nltk.stem import WordNetLemmatizer
lema=WordNetLemmatizer()
from nltk import sent_tokenize
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-Z]',' ',df['Message'][i])
    review=review.lower()
    review=review.split()
    review=[lema.lemmatize(word) for word in review ]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
# train word2vec from scratch
model=gensim.models.Word2Vec(words,vector_size=100)
```

```python
from tqdm import tqdm
```

```python
def avg_word2vec(doc):
    # Filter words to only those in the model's vocabulary
    valid_words = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    
    if valid_words:
        return np.mean(valid_words, axis=0)
    else:
        # Return a zero vector if no valid words are found
        return np.zeros(model.wv.vector_size)
x = []
for i in tqdm(range(len(words))):
    x.append(avg_word2vec(words[i]))
```
